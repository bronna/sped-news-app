```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
```

```python
df1 = pd.read_csv('sped-18-19-clean.csv')
df1.shape
```

```python
df2 = pd.read_csv('oregon_sd_2018_shape_data.csv')
df2.shape
```

```python
df1.name = df1.name.str.replace("RJ", "J")
```

```python
df1.head(15)
```

```python
df2.head()
```

```python
df2.NAME = df2.NAME.str.replace("School District", "SD")
```

```python
df2.head()
```

```python
df_merged = pd.merge(df1, df2, left_on="name", right_on="NAME", indicator=True, how='outer')
```
